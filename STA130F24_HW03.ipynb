````python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

# Set Plotly renderer for JupyterHub (optional if you're using "png" renderer)
pio.renderers.default = 'png'

# Load the penguins dataset
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Create a 3x1 layout (3 rows, 1 column) for the figures of each species
fig = make_subplots(rows=3, cols=1, subplot_titles=penguins['species'].dropna().unique())

# Function to calculate IQR
def calculate_iqr(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    return q1, q3, iqr

# Plot the histogram and add lines/rectangles for each species
for i, species in enumerate(penguins['species'].dropna().unique()):
    # Filter data by species
    species_data = penguins[penguins['species'] == species]['flipper_length_mm'].dropna()

    # Calculate statistics
    mean = species_data.mean()
    median = species_data.median()
    min_val = species_data.min()
    max_val = species_data.max()
    std_dev = species_data.std()
    q1, q3, iqr = calculate_iqr(species_data)
    
    # Create histogram with more bins to reduce gaps
    hist = go.Histogram(x=species_data, name=species, nbinsx=30, opacity=0.75)

    # Add histogram to subplot (set row=i+1 for each species, same column=1)
    fig.add_trace(hist, row=i+1, col=1)
    
    # Add vertical lines for mean and median
    fig.add_vline(x=mean, line=dict(color='blue', dash='dash'), row=i+1, col=1, annotation_text="Mean", annotation_position="top right")
    fig.add_vline(x=median, line=dict(color='green', dash='dot'), row=i+1, col=1, annotation_text="Median", annotation_position="top right")

    # Add rectangles for range, IQR, and 2 std dev
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor='lightgray', opacity=0.3, row=i+1, col=1, annotation_text="Range")
    fig.add_vrect(x0=q1, x1=q3, fillcolor='yellow', opacity=0.4, row=i+1, col=1, annotation_text="IQR")
    fig.add_vrect(x0=mean-2*std_dev, x1=mean+2*std_dev, fillcolor='red', opacity=0.2, row=i+1, col=1, annotation_text="2 Std Dev")

# Update layout with a wider figure size
fig.update_layout(title_text='Histograms of Flipper Length for Each Species', showlegend=False, height=900, width=1200)

# Show plot as static PNG
fig.show(renderer="png")
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset from CSV
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Create a figure with 3 subplots in a row (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18, 8))  # Increased height for longer plots

# Function to calculate IQR
def calculate_iqr(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    return q1, q3, iqr

# Plot KDE for each species and add lines/rectangles for mean, median, range, IQR, and 2 std dev
species_list = df['species'].dropna().unique()
for i, species in enumerate(species_list):
    # Filter data for each species
    species_data = df[df['species'] == species]['flipper_length_mm'].dropna()

    # Calculate statistics
    mean = species_data.mean()
    median = species_data.median()
    min_val = species_data.min()
    max_val = species_data.max()
    std_dev = species_data.std()
    q1, q3, iqr = calculate_iqr(species_data)
    
    # Create KDE plot
    sns.kdeplot(species_data, ax=axes[i], fill=True, color='skyblue')
    
    # Add vertical lines for mean and median
    axes[i].axvline(mean, color='red', linestyle='--', label='Mean')
    axes[i].axvline(median, color='blue', linestyle='-.', label='Median')
    
    # Add shaded area for range
    axes[i].axvspan(min_val, max_val, color='gray', alpha=0.2, label='Range')
    
    # Add shaded area for IQR
    axes[i].axvspan(q1, q3, color='yellow', alpha=0.3, label='IQR')
    
    # Add shaded area for two standard deviations from the mean
    axes[i].axvspan(mean - 2*std_dev, mean + 2*std_dev, color='purple', alpha=0.2, label='2 Std Dev')
    
    # Add titles and labels
    axes[i].set_title(f'{species} KDE of Flipper Length')
    axes[i].set_xlabel('Flipper Length (mm)')
    axes[i].set_ylabel('Density')

    # Add a legend
    axes[i].legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
````

For right skewed, data tended to the right and resulting that mean is greater than median. The median is less affected by such large changes.

From the graphs below, we are able to see that for a right skewed distribution, that the mean is less than the medium, and similarly for left skewed, the mean becomes slightly larger than the median.

````python
#Right skewed graph
from scipy import stats
import pandas as pd
import numpy as np
import plotly.express as px  # Assuming you meant to use Plotly Express
import plotly.graph_objects as go

# Generate a sample from a Gamma distribution
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)

# Calculate the mean and median
mean_sample1 = sample1.mean()
median_sample1 = np.quantile(sample1, 0.5)  # median

# Create the histogram
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Gamma Distribution with Mean and Median")

# Add vertical lines for mean and median
fig1.add_vline(x=mean_sample1, line_dash="dash", line_color="green", annotation_text="Mean", annotation_position="top left")
fig1.add_vline(x=median_sample1, line_dash="dash", line_color="blue", annotation_text="Median", annotation_position="top right")

# Show the plot
fig1.show(renderer="png") 
````

````python
from scipy import stats
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# Generate a sample from a left-skewed distribution by negating a Gamma distribution
sample1 = -stats.gamma(a=2, scale=2).rvs(size=1000)

# Calculate the mean and median
mean_sample1 = sample1.mean()
median_sample1 = np.quantile(sample1, 0.5)  # median

# Create the histogram
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Left-Skewed Gamma Distribution with Mean and Median")

# Add vertical lines for mean and median
fig1.add_vline(x=mean_sample1, line_dash="dash", line_color="green", annotation_text="Mean", annotation_position="top left")
fig1.add_vline(x=median_sample1, line_dash="dash", line_color="blue", annotation_text="Median", annotation_position="top right")

# Show the plot
fig1.show(renderer="png")  # This is useful for submissions or platforms like GitHub and MarkUs
````
